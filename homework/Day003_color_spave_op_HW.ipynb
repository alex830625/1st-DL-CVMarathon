```python
import cv2
import numpy as np


#調整圖片飽和度
#圖片路徑和讀圖

img_path='/Users/alex8/Documents/GitHub/1st-DL-CVMarathon/homework/img/lena.png'
img=cv2.imread(img_path,cv2.IMREAD_COLOR)

#把圖片轉成HSV的色彩空間
img_hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

#因為要轉成百分比，所以有小數點
img_hsv_down=img_hsv.astype('float32')

#編碼是0到255，因此先轉換成0到1之間的數字，再減百分之20
img_hsv_down[:,:,1]=img_hsv_down[:,:,1]/255 - 0.2

#最小值會在0到1之間，所以為了防止數值小於0，設定最小值為0
img_hsv_down[img_hsv_down[:,:,1] < 0] = 0

#把計算好的數值套回去原本的大小
img_hsv_down[:,:,1]=img_hsv_down[:,:,1]*255

#將格式轉換為uint8
img_hsv_down = img_hsv_down.astype('uint8')

#最後轉回BGR色彩空間
img_hsv_down = cv2.cvtColor(img_hsv_down,cv2.COLOR_HSV2BGR)

img_hsv_up=img_hsv.astype('float32')
img_hsv_up[:,:,1]=img_hsv_up[:,:,1]/255 + 0.2
img_hsv_up[img_hsv_up[:,:,1] > 1 ] = 1
img_hsv_up[:,:,1]=img_hsv_up[:,:,1]*255
img_hsv_up = img_hsv_up.astype('uint8')
img_hsv_up = cv2.cvtColor(img_hsv_up,cv2.COLOR_HSV2BGR)


#把圖片合在一起並顯示圖片
img_hsv_change = np.hstack((img,img_hsv_down,img_hsv_up))
cv2.imshow('img_hsv', img_hsv_change)



#按任意鍵關閉視窗
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python

import cv2
import numpy as np

#直方圖均衡
img_path='/Users/alex8/Documents/GitHub/1st-DL-CVMarathon/homework/img/lena.png'
img=cv2.imread(img_path,cv2.IMREAD_COLOR)
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img_hsv= cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

#灰階圖片直方圖均衡
img_gray_equal =cv2.equalizeHist(img_gray)

#每個channel 個別做直方圖均衡
#img[..., 1]表⽰示取得第 1 個 channel 的row*column 的值
equalHist_by_channel = [img[..., 0], img[..., 1], img[..., 2]]
equalHist_by_channel = [cv2.equalizeHist(i) for i in equalHist_by_channel]
# 組合經過直方圖均衡的每個 channel
#npstack是"Join a sequence of arrays along a new axis.
# axis=-1表示是最後一個dimension
img_bgr_equal = np.stack(equalHist_by_channel, axis = -1)

#轉換color space 後只對其中一個channel做直方圖均衡

img_hsv_onechannel = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hsv_onechannel[:,:,2] = cv2.equalizeHist(img_hsv_onechannel[:,:,2])


#組合+顯示圖片
img_gray_equalHist = np.hstack((img_gray,img_gray_equal))
img_color_equalHist = np.hstack((img_bgr_equal, img_hsv_onechannel))


cv2.imshow('img_gray_equalHist',img_gray_equalHist)
cv2.imshow('img_color_equalHist',img_color_equalHist)

#按任意鍵關閉視窗
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

import cv2
import numpy as np

img_path='/Users/Alex/Documents/GitHub/1st-DL-CVMarathon/homework/img/lena.png'
img=cv2.imread(img_path,cv2.IMREAD_COLOR)

# alpha: 控制對比度 (1.0~3.0)
# beta: 控制明亮度 (0~255)


add_contrast = cv2.convertScaleAbs(img, alpha=2.0, beta=0)
add_lightness = cv2.convertScaleAbs(img, alpha=1.0, beta=50)


#組合+顯示圖片
img_gray_equalHist = np.hstack((img,add_contrast,add_lightness))
cv2.imshow('test',img_gray_equalHist)


#按任意鍵關閉視窗
cv2.waitKey(0)
cv2.destroyAllWindows()













```
